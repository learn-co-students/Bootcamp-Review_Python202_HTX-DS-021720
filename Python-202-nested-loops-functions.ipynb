## While, Break, and Continue

What does a `while` loop look like in Python? Let's look at an example:

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

What are `break` and `continue`? How they different?

Run the following code:

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print("foo")
    continue
  print(i)
```

How does the code above work?

Now run this code:

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print("foo")
    break
  print(i)
```
